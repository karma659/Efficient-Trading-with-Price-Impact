```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the provided dataset
data_path = "/content/sample_data/merged_data.csv"
data = pd.read_csv(data_path)

# Quick inspection of the data
data.head(), data.describe()
```

```python
data.info()
```

```python
# Parameters for the OW model

# Calculate time_steps
time_steps = len(data)

# Convert 'ts_event' to datetime and calculate dt
data['ts_event'] = pd.to_datetime(data['ts_event'])
data['dt'] = data['ts_event'].diff().dt.total_seconds()
dt = data['dt'].dropna().mean()  # Average time interval

# calculate lambda_ow
lambda_ow = np.abs(data['Signed Volume']).mean()

# Calculate price changes and autocorrelation
data['price_changes'] = data['price'].diff()
price_changes = data['price_changes'].dropna()
autocorr = price_changes.autocorr()

# Calculate beta_ow
beta_ow = price_changes.var() / lambda_ow

# Output results
print(f"time_steps: {time_steps}, dt: {dt}, lambda_ow: {lambda_ow}, beta_ow: {beta_ow}")
```

```python

# Simulate inventory (Q_t) based on Signed Volume and compute price impact
inventory_ow = np.cumsum(data['Signed Volume'].values) * dt  # Cumulative inventory
impact_ow = lambda_ow * inventory_ow                        # Linear price impact

```

```python
# Compute decayed impact (J_t) over time
decayed_impact_ow = np.zeros(time_steps)
for t in range(1, time_steps):
    decayed_impact_ow[t] = decayed_impact_ow[t-1] * np.exp(-beta_ow * dt) + impact_ow[t]
```

```python
# Visualize the distribution of price impact
plt.figure(figsize=(10, 6))
plt.hist(decayed_impact_ow, bins=50, alpha=0.7, label='Decayed Impact (Linear OW)', color='blue')
plt.xlabel("Price Impact")
plt.ylabel("Frequency")
plt.title("Distribution of Price Impact: Linear OW Model")
plt.legend()
plt.show()
```

```python

# Load and preprocess data
data['ts_event'] = pd.to_datetime(data['ts_event'])
data['price_changes'] = data['price'].diff()

# 1. Calculate lambda_afs (mean price change per unit signed volume)
mean_signed_volume = data['Signed Volume'].mean()
mean_price_change = data['price_changes'].mean()
lambda_afs = mean_price_change / mean_signed_volume

# Drop rows with NaN or zero values in price_changes or Signed Volume
data = data.dropna(subset=['price_changes', 'Signed Volume'])
data = data[(data['Signed Volume'] != 0) & (data['price_changes'] != 0)]

# Recalculate log transformations
log_signed_volume = np.log(np.abs(data['Signed Volume']))
log_price_changes = np.log(np.abs(data['price_changes']))

# Ensure no infinities or NaN values in log data
valid_indices = ~(np.isinf(log_signed_volume) | np.isinf(log_price_changes) |
                  np.isnan(log_signed_volume) | np.isnan(log_price_changes))
log_signed_volume = log_signed_volume[valid_indices]
log_price_changes = log_price_changes[valid_indices]

# Recalculate p_afs using log-log regression
p_afs = np.polyfit(log_signed_volume, log_price_changes, 1)[0]  # Slope of log-log regression

# 3. Calculate beta_afs (autocorrelation decay parameter based on price_changes)
autocorr = data['price_changes'].autocorr()
beta_afs = 1 - np.abs(autocorr)  # Simplified decay estimate

# Return results
results = {
    'lambda_afs': lambda_afs,
    'p_afs': p_afs,
    'beta_afs': beta_afs,
    'log_signed_volume_length': len(log_signed_volume),
    'log_price_changes_length': len(log_price_changes)
}

results
```

```python
# Compute nonlinear price impact: I_t = lambda * |Q_t|^p
impact_afs = lambda_afs * np.sign(inventory_ow) * (np.abs(inventory_ow) ** p_afs)

# Compute decayed impact (J_t) over time
decayed_impact_afs = np.zeros(time_steps)
for t in range(1, time_steps):
    decayed_impact_afs[t] = decayed_impact_afs[t-1] * np.exp(-beta_afs * dt) + impact_afs[t]

# Visualize the distribution of price impact
plt.figure(figsize=(10, 6))
plt.hist(decayed_impact_afs, bins=50, alpha=0.7, label='Decayed Impact (Nonlinear AFS)', color='green')
plt.xlabel("Price Impact")
plt.ylabel("Frequency")
plt.title("Distribution of Price Impact: Nonlinear AFS Model")
plt.legend()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Reload the dataset
file_path = '/content/sample_data/merged_data.csv'
data = pd.read_csv(file_path)

# Inspect the dataset
print("Dataset columns:", data.columns)

# Extract relevant features
signal_column = 'Signed Volume'
price_column = 'mid_price'

# Handle missing or invalid data
data = data.dropna(subset=[signal_column, price_column])
data = data[data[signal_column] != 0]

# Prepare data for training
f_n = data[signal_column].values
J_n = data[price_column].values

# Normalize inputs
f_n = f_n / np.max(np.abs(f_n))
J_n = J_n / np.max(np.abs(J_n))
inputs = np.column_stack((f_n[:-1], J_n[:-1]))
targets = f_n[1:]

# Define neural network architectures
def create_network(architecture="simple"):
    if architecture == "simple":
        model = models.Sequential([
            layers.Dense(128, activation="relu"),
            layers.Dense(32, activation="relu"),
            layers.Dense(8, activation="relu"),
            layers.Dense(1)
        ])
    elif architecture == "power":
        model = models.Sequential([
            layers.Dense(128, activation="relu"),
            layers.Dense(32, activation="relu"),
            layers.Dense(8, activation="relu"),
            layers.Dense(1, activation="sigmoid")
        ])
    elif architecture == "linear":
        model = models.Sequential([layers.Dense(1)])
    return model

# Train the models
def train_model(architecture, inputs, targets, epochs=50):
    model = create_network(architecture)
    model.compile(optimizer=optimizers.Adam(learning_rate=1e-4), loss="mse")
    history = model.fit(inputs, targets, epochs=epochs, batch_size=128, verbose=1, validation_split=0.2)
    return model, history

# Train and compare architectures
architectures = ["simple", "power", "linear"]
histories = {}

for arch in architectures:
    print(f"Training {arch} architecture...")
    _, history = train_model(arch, inputs, targets)
    histories[arch] = history

# Visualize training losses
plt.figure(figsize=(10, 6))
for arch in architectures:
    plt.plot(histories[arch].history['loss'], label=f'{arch} Train Loss')
    plt.plot(histories[arch].history['val_loss'], label=f'{arch} Validation Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss for Different Architectures")
plt.legend()
plt.grid(True)
plt.show()
```
